```python

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "cell_phones_total.csv"
df = pd.read_csv(file_path)

# Extract years as columns (excluding 'country')
years = df.columns[1:]  
df_numeric = df[years].apply(pd.to_numeric, errors='coerce')

# Compute global average usage per year
avg_usage = df_numeric.mean(skipna=True)

# Plot the line chart with Year on Y-axis and Cell Phones per Person on X-axis
plt.figure(figsize=(8, 10))
plt.plot(avg_usage.values, avg_usage.index, marker='o', linestyle='-', color='b')

plt.xlabel("Number of Cell Phones per 100 People")
plt.ylabel("Year")
plt.title("Rise in Cell Phone Usage Over the Years")
plt.grid()

# Show the plot
plt.show()

```
