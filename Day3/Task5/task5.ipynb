```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Load datasets
income_file = "mincpcap_cppp.csv"
life_expectancy_file = "lex.csv"

income_df = pd.read_csv(income_file)
life_expectancy_df = pd.read_csv(life_expectancy_file)

# Convert data to numeric format
years = income_df.columns[1:]
income_df[years] = income_df[years].apply(pd.to_numeric, errors='coerce')
life_expectancy_df[years] = life_expectancy_df[years].apply(pd.to_numeric, errors='coerce')

# Melt datasets
income_melted = income_df.melt(id_vars=['country'], var_name='year', value_name='income')
life_expectancy_melted = life_expectancy_df.melt(id_vars=['country'], var_name='year', value_name='life_expectancy')

# Merge datasets
merged_df = pd.merge(income_melted, life_expectancy_melted, on=['country', 'year'])

# Select a static year
year_selected = '2000'  # Change this value to see different years
filtered_df = merged_df[merged_df['year'] == year_selected]

# Plot bubble chart
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=filtered_df,
    x='income',
    y='life_expectancy',
    size='income',  # Bubble size based on income
    hue='country',  # Different colors for countries
    palette='tab10',
    alpha=0.6,
    sizes=(10, 1000)
)

plt.xscale('log')  # Log scale for better visualization
plt.xlabel("Income per Capita (Log Scale)")
plt.ylabel("Life Expectancy")
plt.title(f"Life Expectancy vs Income (Year: {year_selected})")
plt.legend([], [], frameon=False)  # Hide automatic legend
plt.grid()
plt.show()
```
