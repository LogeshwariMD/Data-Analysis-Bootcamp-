```python
import pandas as pd
import matplotlib.pyplot as plt



```

```python
# Load the dataset
file_path = "mincpcap_cppp.csv"  # Change path if needed
df = pd.read_csv(file_path)


```

```python
# Transpose the dataset (excluding the 'country' column) and compute average income per year
df_numeric = df.drop(columns=["country"]).astype(float)  # Convert to float for calculations
average_income = df_numeric.mean(axis=0)  # Compute mean across countries

```

```python
# Convert index to integers (years)
years = average_income.index.astype(int)


```

```python
# Plotting
plt.figure(figsize=(8, 6))
plt.plot(average_income, years, marker='o', linestyle='-')

```

```python
# Labels and title
plt.xlabel("Average Income")
plt.ylabel("Year")
plt.title("Average Income Trend Over the Years")
plt.grid(True)

# Show the plot
plt.show()
```
