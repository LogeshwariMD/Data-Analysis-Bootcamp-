```
import matplotlib.pyplot as plt
import pandas as pd

# Sample data (replace with full dataset)
data = {
    "year": [1960, 1970, 1980, 1990, 2000, 2010, 2020],
    "cell_phones_total": [0, 0, 0, 0, 0, 36.2, 58.2],
    "population": [9040000, 11300000, 13200000, 12000000, 20100000, 28300000, 39100000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate Cell Phones Per Person
df["cell_phones_per_person"] = df["cell_phones_total"] / df["population"]

# Plot the graph
plt.figure(figsize=(8, 5))
plt.plot(df["year"], df["cell_phones_total"], marker='o', linestyle='-', label="Total Cell Phones")
plt.plot(df["year"], df["cell_phones_per_person"], marker='s', linestyle='--', label="Cell Phones Per Person")

# Labels and Title
plt.xlabel("Year")
plt.ylabel("Count")
plt.title("Rise of Cell Phones Over Time")
plt.legend()
plt.grid(True)

# Show the graph
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName("DatabricksChart").getOrCreate()
```

```
years = list(range(1960, 2024))
avg_cell_phones_per_person = [1] * (len(years) - 1) + [1.0]  # Keeping 1 for all years, last value as 1.0

df = pd.DataFrame({"Year": years, "Avg_Cell_Phones_Per_Person": avg_cell_phones_per_person})
```

```
spark_df = spark.createDataFrame(df)
```

```
pandas_df = spark_df.toPandas()
```

```
plt.figure(figsize=(12, 6))
sns.lineplot(data=pandas_df, x="Year", y="Avg_Cell_Phones_Per_Person", marker="o", color="blue")
```
