```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS

# Load the dataset
file_path = "IPL_Match_Highlights_Commentary.csv"
df = pd.read_csv(file_path)

# Check column names
df.columns

# Assuming the commentary is in a column named 'Commentary'
if 'Commentary' in df.columns:
    text = " ".join(comment for comment in df['Commentary'].astype(str))
else:
    raise ValueError("Column 'Commentary' not found in dataset")

# Define stopwords
stopwords = set(STOPWORDS)

# Generate Word Cloud
wordcloud = WordCloud(width=800, height=400, background_color='white', stopwords=stopwords, colormap='viridis').generate(text)

# Display the word cloud
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
