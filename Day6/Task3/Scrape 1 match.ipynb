```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# Setup WebDriver
options = webdriver.ChromeOptions()
options.add_argument("--headless")  # Run in headless mode (no UI)
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64)")

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# ESPN Cricinfo Ball-by-Ball Commentary URL
url = "https://www.espncricinfo.com/series/indian-premier-league-2022-1298423/gujarat-titans-vs-rajasthan-royals-final-1312200/ball-by-ball-commentary"
driver.get(url)
time.sleep(5)  # Allow page to load

# Scroll to load more commentary
for _ in range(10):  # Adjust if necessary
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)

# Find all ball-by-ball commentary elements
balls = driver.find_elements(By.XPATH, "//div[contains(@class, 'ds-text-tight-m')]")

# List to store structured data
data = []

# Extract data for each ball
for ball in balls:
    try:
        # Extract over number
        over_element = ball.find_element(By.XPATH, ".//span[contains(@class, 'ds-text-tight-s')]")
        over = over_element.text.strip() if over_element else "N/A"
        
        # Extract commentary text
        commentary_text = ball.text.strip()
        commentary_parts = commentary_text.split(", ")

        # Extract bowler and batter names
        bowler_batter = commentary_parts[0].split(" to ")
        if len(bowler_batter) == 2:
            bowler = bowler_batter[0].strip()
            batter = bowler_batter[1].strip()
        else:
            bowler = "Unknown"
            batter = "Unknown"

        # Identify ball type and shot type
        ball_type = "Normal"
        shot_type = "None"
        runs = "0"
        speed = "N/A"

        for part in commentary_parts:
            if "wide" in part.lower():
                ball_type = "Wide"
            elif "no ball" in part.lower():
                ball_type = "No Ball"
            elif "out" in part.lower():
                ball_type = "Wicket"
            elif "dot ball" in part.lower():
                ball_type = "Dot Ball"
            elif "four" in part.lower():
                shot_type = "Boundary"
                runs = "4"
            elif "six" in part.lower():
                shot_type = "Six"
                runs = "6"
            elif "single" in part.lower():
                shot_type = "Single"
                runs = "1"
            elif "double" in part.lower():
                shot_type = "Double"
                runs = "2"

            # Extract speed if available
            if "km/h" in part:
                speed = part.strip()

            # Extract runs scored (if last part contains numbers)
            if part.strip().isdigit():
                runs = part.strip()

        # Append structured data
        data.append([over, bowler, batter, ball_type, shot_type, speed, runs])
    
    except Exception as e:
        print(f"Error: {e}")

# Close browser
driver.quit()

# Convert to DataFrame
df = pd.DataFrame(data, columns=["Over", "Bowler", "Batter", "Ball Type", "Shot Type", "Speed (km/h)", "Runs"])

# Save to CSV
df.to_csv("GT_vs_RR_commentary_structured.csv", index=False)

# Print first few rows
print(df.head())


```
