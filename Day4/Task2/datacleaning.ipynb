```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


```

```python
file_path = "validation.csv"
df = pd.read_csv(file_path)
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
print("Categorical Columns:")
print(categorical_cols)

```

```python
bias_columns = ['Gender', 'SeniorCitizen', 'PaymentMethod', 'Contract']  
```

```python
plt.figure(figsize=(12, 8))
for i, col in enumerate(bias_columns, 1):
    plt.subplot(2, 2, i)
    sns.countplot(x=df[col], palette='coolwarm')
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
bias_info = {}
for col in bias_columns:
    bias_info[col] = df[col].value_counts(normalize=True) * 100
    
```

```python
print("Potential Bias in Dataset:")
for col, values in bias_info.items():
    print(f"\n{col} Distribution:")
    print(values)
```
